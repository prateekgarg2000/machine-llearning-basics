```python
def calmc1(data,learning_rate,m):
    n=len(data)
    l=len(data[0])
    mslope=[0 for i in range(l)]
    for i in range(n):
        y=data[i][l-1]
        for j in range(l-1):
            y-=m[j]*data[i][j]
        y-=m[l-1]
        for j in range(l-1):
            mslope[j]+=(-2/n)*(y)*data[i][j]
        mslope[l-1]+=(-2/n)*y
    for j in range(l):
        m[j]=m[j]-learning_rate*mslope[j]
    return m
def gd1(data, learning_rate, iterations,m):
    l=len(data[0])
    for i in range(iterations):
        m=calmc1(data,learning_rate,m)
        if(i%10==0):
            print(i,cost(data,m))
    return m
```

```python
def calmc(data,learning_rate,m):
    n=len(data)
    l=len(data[0])
    mslope=[0 for i in range(l)]
    for i in range(n):
        y=data[i][l-1]
        for j in range(l-1):
            y-=m[j]*data[i][j]
        y-=m[l-1]
        for j in range(l-1):
            mslope[j]+=(-2/n)*(y)*data[i][j]
        mslope[l-1]+=(-2/n)*y
        for j in range(l):
            m[j]=m[j]-learning_rate*mslope[j]
    return m
```

```python
def gd(data, learning_rate, iterations,m):
    l=len(data[0])
    for i in range(iterations):
        m=calmc(data,learning_rate,m)
#         if(i%4==0):
        print(i,cost(data,m))
    return m
```

```python
def cost(data,m):
    n=len(data)
    l=len(data[0])
    t_cost=0
    for i in range(n):
        x=0
        for j in range(l-1):
            x+=data[i,j]*m[j]
        y=data[i,l-1]
        t_cost+=(1/n)*((y-x-m[l-1])**2)
    return t_cost
```
